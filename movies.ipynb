```python
import numpy as np
import pandas as pd
```

```python
movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head(1)
```

```python
credits.head(1)
```

```python
movies=movies.merge(credits, on='title')
```

```python
movies.columns
```

```python
movies=movies[['movie_id','genres','overview','keywords','title','cast','crew']]
```

```python
movies.isna().sum() 
```

```python
movies.dropna(inplace=True)
```

```python
import ast
```

```python
def converter(obj):
  l=[]
  for i in ast.literal_eval(obj):
    l.append(i['name'])
  return l
```

```python
movies.genres=movies.genres.apply(converter)
```

```python
movies.genres[0]
```

```python
movies.keywords
```

```python
movies.keywords=movies.keywords.apply(converter)
```

```python
movies.keywords[0]
```

```python
movies.cast[0]
```

```python
def converter2(obj):
  l=[]
  add=0
  for i in ast.literal_eval(obj):
    if add <= 3:
      l.append(i['name'])
      add = add + 1
    else:
      break
  return l
```

```python
movies.cast=movies.cast.apply(converter2)
```

```python
movies.cast[0]
```

```python
movies.head(1)
```

```python
movies.crew[0]
```

```python
def director(obj):
  l=[]
  for i in ast.literal_eval(obj):
    if i['job']== 'Director':
      l.append(i['name'])
      break
  return l
```

```python
movies.crew=movies.crew.apply(director)
```

```python
movies.crew[0]
```

```python
def Converter3(string):
    l = list(string.split(" "))
    return l
```

```python
movies.overview=movies.overview.apply(Converter3)
```

```python
movies.head(1)
```

```python
movies.genres=movies.genres.apply(lambda x: [i.replace(' ','') for i in x])
```

```python
movies.genres[0]
```

```python
movies.keywords=movies.keywords.apply(lambda x: [i.replace(' ','') for i in x])
movies.cast=movies.cast.apply(lambda x: [i.replace(' ','') for i in x])
movies.crew=movies.crew.apply(lambda x: [i.replace(' ','') for i in x])
```

```python
movies.head(1)
```

```python
movies['information']=movies['overview'] + movies['genres'] + movies['cast'] + movies['keywords'] + movies['crew']
```

```python
movies.head(1)
```

```python
df= movies[['movie_id','title','information']]
```

```python
df.head(1)
```

```python
df.information = df.information.apply(lambda x: ' '.join(x))
```

```python
df.information
```

```python
df.head(1)
```

```python
df.information = df.information.apply(lambda x: x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features=5000, stop_words='english')
```

```python
vector = cv.fit_transform(df['information']).toarray()
```

```python
vector.shape
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stemming(text):
  l=[]
  for i in text.split():
    l.append(ps.stem(i))
  return ' '.join(l)
```

```python
df.information=df.information.apply(stemming)
```

```python
df.information
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similar=cosine_similarity(vector)
```

```python
similar[1]
```

```python
df.title[1:20]
```

```python
def recommended(movie):
  index=df[df['title']== movie].index[0]
  distance= similar[index]
  movie_list=sorted(list(enumerate(distance)),reverse=True, key=lambda x: x[1])[1:6]
  for i in movie_list:
    print(df.title[i[0]])
```

```python
recommended("The Avengers")
```
